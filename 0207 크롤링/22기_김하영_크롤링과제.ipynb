## 1번 과제

```python
import requests
from bs4 import BeautifulSoup
```

```python
#전체 html 저장

url = 'https://www.yonsei.ac.kr/sc/support/calendar.jsp'

res = requests.get(url)
print(res)
res.text
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
#학사일정 html 저장
html = soup.find_all({'th','td'})
len(html)
html
```

```python
#텍스트 추출 및 공백문자 제거

text = []

for h in html:
    t = h.get_text().replace('\n','').replace('\t','').replace('\xa0','')
    text.append(t)
```

```python
#2,3월 학사일정만 저장

Feb = []
Mar = []

for t in text:
    if '03 MAR' == t:
        i_Mar = text.index(t)
        Feb = text[1:i_Mar]
    elif '04 APR' == t:
        i_Apr = text.index(t)
        Mar = text[i_Mar + 1:i_Apr]
        break

print(Feb)
print(Mar)
```

```python
#Column은 월(Month), 일(Day), 학시 일정(Event)

Month = []
Day = []
Event = []

for f in range(0,len(Feb),2):
    Month.append(2)
    Day.append(Feb[f])
    Event.append(Feb[f+1])

for m in range(0,len(Mar),2):
    Month.append(3)
    Day.append(Mar[m])
    Event.append(Mar[m+1])
    
print(Month,Day,Event)
```

```python
#Data Frame 형식으로 저장

import pandas as pd

dic = {'Month':Month, 'Day': Day, 'Event': Event}
df = pd.DataFrame(dic)
df
```

## 2번 과제

```python
import selenium
from selenium import webdriver

driver = webdriver.Chrome()
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# 해당 URL로 이동
driver.get('https://www.naver.com/') 

# 웹 페이지 요소 로드를 기다림
driver.implicitly_wait(5)
```

```python
# 현재 열려있는 페이지 url 찾기
url = driver.current_url
url
```

```python
# XPATH를 이용하여 검색창에 해당하는 요소 찾기
search = driver.find_element(By.XPATH, '//*[@id="query"]')
```

```python
# 검색어 입력
search.send_keys('고려대학교')
```

```python
# Keys.RETURN = Enter 입력
search.send_keys(Keys.RETURN)
```

```python
#이미지 url 저장
img = driver.find_element(By.XPATH, '//*[@id="main_pack"]/div[2]/div[2]/div[1]/div/div[2]/a/img')
img_url = img.get_attribute('src')
```
